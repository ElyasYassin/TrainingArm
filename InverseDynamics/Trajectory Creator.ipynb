### Check the number of joints of the given MyoSuite task's model

```python


env = gym.make('CenterReachOut-v0', normalize_act = False) #Change the task depending on what model/environment you want to use
env.reset()

env.env.init_qpos[:] = np.zeros(len(env.env.init_qpos),)
mjcModel = env.env.sim.model

joints = ['time']

print("\nJoints:")
for i in range(mjcModel.njnt):
    joints.append(mjcModel.joint(i).name)
    print([mjcModel.joint(i).name])
     
print(joints)
print(len(joints))  
```

```python
import myosuite
import gym
import pickle

movements = []

#Import Policy
policy = "policy_2000.pickle"
pi = pickle.load(open(policy, 'rb'))


env = gym.make('CenterReachOut-v0', normalize_act = True)
env.reset()
for _ in range(20):
    movements.append(env.obs_dict['hand_qpos'])
    env.mj_render()
    o = env.get_obs()
    # get the next action from the policy
    action, _ = pi.get_action(o)
    env.step(action) # take a random action
    
```

```python
filename = 'traj.csv'

time_values = [i * 0.002 for i in range(int(10 / 0.5) + 1)] #10 seconds / increment of 0.5seconds

print(len(time_values))

with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    
    # Write column headers
    writer.writerow(joints)
    
    time = 0
    
    x = 0
    
    for time_val in time_values:
        x += 1
        row = [time_val]  # Start with the time value
        # For each column (except 'time'), add a placeholder value (0.0 in this case)
        for _ in range(len(joints)-1):
            print(len(joints))
            print("x: ", x)
            print("_: ",_)
            try:
                row.append(movements[x][_])
            except IndexError:
                row.append(0.0)

        # Write the row to the CSV file
        writer.writerow(row)
```
